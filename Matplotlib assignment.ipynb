```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
file_path ='jumta.csv'
df=pd.read_csv(file_path)
df
```

```python
plt.figure(figsize=(10, 10))
plt.plot(df['Month'], df['Revenue'], marker='o', linestyle='-', color='b')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.title('Monthly Revenue Over Two Years')
plt.grid(True)
plt.savefig('monthly_revenue.png', dpi=300)
plt.xticks(rotation=90)
plt.show
```

```python
plt.figure(figsize=(10, 10))
plt.bar(df['Month'], df['Orders'], color='g')
plt.xlabel('Month')
plt.ylabel('Total Number of Orders')
plt.title('Total Number of Orders Per Month')
plt.xticks(rotation=90)
plt.savefig('Total_orders', dpi=300)
plt.show
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(df['Orders'], df['Average_Order_Value'], color='y')
plt.xlabel('Number of Orders')
plt.ylabel('Average Order Value')
plt.title('Number of Orders vs. Average Order Value')
plt.savefig('orders_vs_avg_value', dpi=300)
plt.show
```

```python
fig, axes = plt.subplots(nrows=3, figsize=(10, 20))
axes[0].plot(df['Month'], df['Revenue'], marker='o', linestyle='-', color='b')
axes[0].set_xlabel('Month')
axes[0].set_ylabel('Revenue')
axes[0].set_title('Monthly Revenue Over Two Years')
axes[0].grid(True)
plt.xticks(rotation=90)
axes[1].bar(df['Month'], df['Orders'], color='g')
axes[1].set_xlabel('Month')
axes[1].set_ylabel('Total Number of Orders')
axes[1].set_title('Total Number of Orders Per Month')
plt.xticks(rotation=90)
axes[2].scatter(df['Orders'], df['Average_Order_Value'], color='y')
axes[2].set_xlabel('Number of Orders')
axes[2].set_ylabel('Average Order Value')
axes[2].set_title('Number of Orders vs. Average Order Value')
plt.xticks(rotation=90)
plt.show
```
